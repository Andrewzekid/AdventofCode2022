```python
#Create a dictionary for what each of the opponent's responses beat,ties with and loses
#A for Rock, B for Paper, and C for Scissors.
opp_beats = {"A":"Z","B":"X","C":"Y"}
opp_ties = {k1:k2 for k1,k2 in zip(opp_beats.keys(),rsp_dict.keys())}
opp_loses = {"A":"Y","B":"Z","C":"X"}
```

```python
def calc_score(shape,outcome):
  """
  Calculates the total score for the round
  Parameters:
    shape: Shape you selected
    outcome: the score for the outcome of the round
  """
  return rsp_dict[shape] + outcome
```

```python
def calculate_win_loss(opp_shape,shape):
  """
  Function that returns true if you won, returns false if you lost, or returns None in the event of a draw. 
  Parameters:
    opp_shape: Shape that your opponent plays
    shape: shape that you play
  """
  #Find the shapes that the opponents shape beats, wins, and draws with
  if opp_beats[opp_shape] == shape:
    #Opponent beats you
    return False
  elif opp_loses[opp_shape] == shape:
    #Opponent loses to you
    return True
  elif opp_ties[opp_shape] == shape:
    #Opponent draws to you
    return None
  else:
    #Error
    raise ValueError("Are you sure you provided a valid shape?")
```

"Anyway, the second column says how the round needs to end: X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win. Good luck!"

```python
def find_your_shape(opps_shape,required_outcome):
  """
  Function that returns the shape that you must play if you want to achieve the required outcome
  Parameters:
    opps_shape: shape that the opponent plays
    required_outcome: the outcome that is required by the stratergy guide
  """
  if required_outcome == "X":
    #You need to lose
    return opp_beats[opps_shape]
  elif required_outcome == "Y":
    #You need to draw
    return opp_ties[opps_shape]
  else:
    #You need to win
    return opp_loses[opps_shape]
```

```python
#Total score for all rouonds
all_round_score = 0

for row in splitted_input:
  #find the opps shape and the outcome you need to obtain
  opp_shape,needed_outcome = find_shapes(row)

  #find the shape you need to play to get that outcome
  your_shape = find_your_shape(opp_shape,needed_outcome)

  #Calculate win or loss or draw
  result = calculate_win_loss(opp_shape,your_shape)

  #Calculate the score for the outcome of the round
  outcome_score = score(result)

  #Calculate the total score for the round
  total_score = calc_score(your_shape,outcome_score)
  all_round_score += total_score

```
