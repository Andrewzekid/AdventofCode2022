```python
def return_details(inp):
  """Function that returns the amount of crates to move, which stack to take them from and which stack to move them to in the form of (crate_amnt,starting_stack,ending_stack)
  """
  pattern = re.compile(r"\d+")
  #get all the matches into a list
  matches = list(re.findall(pattern,inp))
  crate_amnt,starting_stack,ending_stack = matches

  #convert numbers in the form of strings into integers
  return (int(crate_amnt),int(starting_stack),int(ending_stack))  
```

```python
def slice_list(n,my_list):
  """Function that returns both the sliced section and the new list after removing n items from the back of a list in form (sliced_section,new_list)"""
  return (my_list[((n)*-1):],my_list[:((n)*-1)])
```
